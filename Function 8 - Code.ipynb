# 5. Surrogate function

```python
# Convert to PyTorch tensors
X_train = torch.tensor(X, dtype=torch.float32)

if y.ndim == 1:
    y = y.reshape(-1, 1)

y_train = torch.tensor(y, dtype=torch.float32)


class SurrogateNet(nn.Module):
    def __init__(self, input_dim=8):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, 64),
            nn.Tanh(),
            nn.Linear(64, 64),
            nn.Tanh(),
            nn.Linear(64, 64),
            nn.Tanh(),
            nn.Linear(64, 1),
        )

    def forward(self, x):
        return self.net(x)

model = SurrogateNet(input_dim=8)

# Train the surrogate
optimizer = optim.Adam(model.parameters(), lr=1e-3)
loss_fn = nn.MSELoss()

n_epochs = 75000
for epoch in range(n_epochs):
    optimizer.zero_grad()
    y_pred = model(X_train)
    loss = loss_fn(y_pred, y_train)
    loss.backward()
    optimizer.step()
```

# 6. Acquisition function - Gradient Descent

```python
dim = 8
```

```python
def local_optimize_surrogate(model, x0, n_steps=500, lr=0.05):
    x = x0.clone().detach().to("cpu")
    x.requires_grad_(True)

    opt_x = optim.Adam([x], lr=lr)

    for step in range(n_steps):
        opt_x.zero_grad()
        y_pred = model(x)        # shape (1, 1)
        loss = -y_pred           
        loss.backward()
        opt_x.step()

        with torch.no_grad():
            x.clamp_(0.0, 1.0)

    with torch.no_grad():
        y_opt = model(x)

    return x.detach(), y_opt.detach()


n_restarts = 300   # number of different random starting points

best_x = None
best_y = None

for i in range(n_restarts):
    # Random start in [0,1]^dim
    x0 = torch.rand(1, dim, device="cpu")

    x_local, y_local = local_optimize_surrogate(model, x0)

    if best_y is None or y_local.item() > best_y.item():
        best_x = x_local
        best_y = y_local

print("x* (8D) =", best_x.cpu().numpy())
print("surrogate f(x*) =", best_y.item())
```
